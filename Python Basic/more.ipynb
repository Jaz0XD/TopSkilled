```python
# Write a function that checks whether a given string of parentheses is valid. A string is valid if:
# 	•	Every ( has a matching )
# 	•	Every { has a matching }
# 	•	Every [ has a matching ]


def is_valid_parentheses(get):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}
    for char in get:
        if char in mapping:
            top_element = stack.pop() if stack else '#'
            if mapping[char] != top_element:
                return False
        else:
            stack.append(char)
    return not stack

print(is_valid_parentheses("()"))   
print(is_valid_parentheses("()[]{}"))
print(is_valid_parentheses("(]"))    
print(is_valid_parentheses("([)]"))  
print(is_valid_parentheses("{[]}")) 
```

```python
def checker(p):
    stack=[]
    mapping= {')': '(', '}': '{', ']': '['}
    for i in p:
        if i in mapping:
            topElement = stack.pop() if stack else '#'
            if mapping[i] != topElement:
                return False
        else:
            stack.append(i)
    return not stack

print(checker('([)]'))
```

```python
def quiz_game():
    Score = 0  
    Qdict = {
        "What is 10 + 2?": "12",
        "What is the capital city of India?": "Delhi",
        "What is the capital of TN?": "Chennai",
        "What is the color of an Orange?": "orange",
        "What is 90 + 5?": "95",
        "What is the opposite of stand?": "sit",
        "Square of 2?": "4",
        "How many sides in a square?": "4",
        "Opposite of white?": "black",
        "Was the quiz easy?": "yes"
    }

    print('Want to play a Quiz? (Y/N): ', flush=True)
    quiz = input()
    
    if quiz == 'Y':
        for question, answer in Qdict.items():
            print(question, flush=True)
            user_answer = input("Ans: ")
            
            if user_answer == answer:
                print("Correct")
                Score += 1
            else:
                print("Incorrect")
        
        print("Total Score:", Score)
        
    elif quiz == 'N':
        print("Quiz exited.")

    else:
        print("Invalid input")

quiz_game()
```
